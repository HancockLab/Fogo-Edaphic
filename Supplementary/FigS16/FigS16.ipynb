```python
import pandas as pd
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.formula.api import ols
import numpy as np
from scipy.stats import mannwhitneyu
from statsmodels.stats.multicomp import MultiComparison
import matplotlib.patches as mpl_patches
import scipy.stats as stats
import os
import matplotlib.gridspec as gridspec
```

```python
df = pd.read_table('FigS16.csv', sep = "\t")
df.head()
```

```python
elements = ['B','Ca','Cd','Co',
            'Cu','Fe','K','Li','Mg',
            'Mn','Mo','Na','Ni','P',
            'Pb','Rb','S','Sr','Zn']
control=df[df['Sample'] == 'S1-1']
mutant=df[df['Sample'] == 'S1-1-irt1']
Order=['S1-1',
       'S1-1-irt1']
def leg(idx1,idx2):
    h, l = axes[idx1,idx2].get_legend_handles_labels()
    axes[idx1,idx2].legend(h, l)
    axes[idx1,idx2].legend(h[0:3], l[0:3]) 
def plots(element,idx1,idx2):
    sns.boxplot(x=df['Sample'], 
                y=element,
                color='grey',
                data=df,
                order=Order,
                linewidth=1.5,
                ax=axes[idx1,idx2])

    sns.swarmplot(x=df['Sample'],
              y=element, 
              data=df,
              order=Order,
              color="0",
              ax=axes[idx1,idx2])
    handles = [mpl_patches.Rectangle((0, 0), 1, 1, fc="white", ec="white", 
                                 lw=0, alpha=0)] * 2
    labels = []
    r,p = mannwhitneyu(control[element],mutant[element])
    labels.append(r"$P$ = {:.2e}".format(p))
    axes[idx1,idx2].legend(handles, labels, loc='best', fontsize='small', 
          fancybox=True, framealpha=0.7, 
          handlelength=0, handletextpad=0)
    
    axes[idx1,idx2].set_xlabel('')
    axes[idx1,idx2].set_ylabel('')
    axes[idx1,idx2].set_title(element,fontsize=18)


fig,axes = plt.subplots(5,4,sharey=False,figsize=(12,14))
a,b=0,0
# for index, value in enumerate(elements):
for index in range(len(elements)):
    if index % 4 == 0:
        if index == 0:
            plots(elements[index],a,b)
            b+=1
        else:
            a+=1
            b=0
            plots(elements[index],a,b)
            b+=1
    elif index == 18:
        plots(elements[index],a,b)
        axes[a,b+1].axis('off') 
    else:
        plots(elements[index],a,b)
        b+=1

sns.despine(trim=True)
plt.tight_layout()
plt.savefig("FigS16.png", format="png",bbox_inches="tight",facecolor='w',dpi=800)
```
